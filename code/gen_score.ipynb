```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
src = pd.read_csv("../data/rating.csv",index_col=0)
src
```

```python
def obj2int(val):
    return int(val.replace(',', ''))

src = src.applymap(obj2int)
src.info()
```

```python
high_score = pd.DataFrame({"score":src.iloc[:, 7:10].sum(axis=1)/src.iloc[:, :].sum(axis=1)})
high_score['score'] *= 100
high_score.describe()#['score'].idxmax()
#high_score.max()
```

```python
low_score = pd.DataFrame({"score":src.iloc[:,0:3].sum(axis=1)/src.iloc[:, :].sum(axis=1)})
low_score *= 100
#['score'].idxmax()
low_score.describe()
```

```python
median_score = pd.DataFrame({"score":src.iloc[:,3:7].sum(axis=1)/src.iloc[:, :].sum(axis=1)})
median_score['score'] *= 100
median_score.describe()
```

```python
high_score.to_csv(path_or_buf="../data/high_score.csv")
low_score.to_csv(path_or_buf="../data//low_score.csv")
median_score.to_csv(path_or_buf="../data/median_score.csv")
```
